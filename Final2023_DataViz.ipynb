## รวมตารางช่วงเวลา 07/12/2017-19/12/2017  แคชเชียร์คนไหนใช้เวลาทำงานเยอะที่สุดและคนไหนใช้เวลาทำงานน้อยที่สุดและดูยอดการชำระด้วยเครดิตการ์ดและเงินสดทั้งหมด ในปี 2017
* ตอบ ปี 2017 วันที่ 7/12 - 19/12 แคชเชียร์ที่ใช้เวลาทำงานเยอะที่สุด คือ OperatorID 268
* จากตารางพบว่า cash มียอดการชำระอยู่ที่ 2​,112,171.46 us ส่วน card มียอดการชำระอยู่ที่ 2,943,462.16 us เมื่อคิดเป็นเปอร์เซ็นต์จะได้ว่า card = 58.2% cash = 41.8%

````python
df1 = pd.read_csv('/content/gdrive/MyDrive/DataViz23Data/final/POS_operator_logs_20171207-20171219.csv')
df1
````

````python
df2 = pd.read_csv('/content/gdrive/MyDrive/DataViz23Data/final/POS_transactions_20171207-20171219.csv')
df2
````

````python
print(df1.isnull().any())
print(df2.isnull().any())
````

###   นำ ```POS_transactions_20171207-20171219``` เชื่อมตารางกับ```POS_operator_logs_20171207-20171219``` ด้วย ```OperatorID``` เพื่อที่จะรวมข้อมูลในช่วงเวลาเดียวกัน


````python
P2017 = df2.merge(df1, how ='left', left_on='OperatorID',right_on='OperatorID')
print(P2017.shape)
P2017
````

````python
P2017nodrop = P2017.drop_duplicates(subset='TranID_x')
P2017nodrop
````

````python
P2017nodrop.to_csv('P2017.csv', index=False)
````

### เพิ่มคอลัมน์ Payment method ที่แปลงค่าจากคอลัมน์ TNcard,TNcash ที่มีค่าเป็น True,False เปลี่ยเป็นค่า card และ cash

````python
import pandas as pd

# สร้าง DataFrame
P2017 = pd.DataFrame(P2017nodrop)

# สร้างฟังก์ชันสำหรับแปลงค่า
def convert_value(TNcash):
    if TNcash ==True:
        return "cash"
    else:
        return "card"

# ใช้ .apply() เพื่อสร้างคอลัมน์ใหม่ที่แปลงค่าจาก Original_Column
P2017nodrop['Payment method'] = P2017nodrop['TNcash'].apply(convert_value)

print(P2017)
````

````python
P2017.reset_index()
````

````python
P2017.to_csv('P2017new.csv', index=False)
````

## รวมตารางช่วงเวลา 13/02/2019-26/02/2019 เพื่อนำไปเปรียบเทียบรายรับของ 2 ช่วงเวลาใน 1 ปี
* ตอบ ในปี 2019 ในช่วงแรกมียอดขายมากกว่า ช่วงหลัง เกือบเท่าตัวเนื่องจากมีช่วงเวลาเฉลี่ยที่เครื่องแคชเชียร์ว่างน้อยกว่าช่วงหลังเกือบเท่าตัวเช่นกัน

````python
df3 = pd.read_csv('/content/gdrive/MyDrive/DataViz23Data/final/POS_operator_logs_20190213-20190226.csv')
df3
````

````python
df4 = pd.read_csv('/content/gdrive/MyDrive/DataViz23Data/final/POS_transactions_20190213-20190226.csv')
df4
````

````python
print(df3.isnull().any())
print(df4.isnull().any())
````

###   นำ ```POS_operator_logs_20190213-20190226``` เชื่อมตารางกับ```POS_transactions_20190213-20190226``` ด้วย ```OperatorID``` เพื่อที่จะรวมข้อมูลในช่วงเวลาเดียวกัน


````python
P2019F = df4.merge(df3, how ='left', left_on='OperatorID',right_on='OperatorID')
print(P2019F.shape)
P2019F
````

````python
P2019Fnodrop = P2019F.drop_duplicates(subset='TranID_x')
P2019Fnodrop
````

````python
P2019Fnodrop.groupby('OperatorID')['Amount'].sum()
````

### เพิ่มคอลัมน์ Payment method ที่แปลงค่าจากคอลัมน์ TNcard,TNcash ที่มีค่าเป็น True,False เปลี่ยเป็นค่า card และ cash

````python
P2019F = pd.DataFrame(P2019Fnodrop)

def combine_values(column):
    if column['TNcash'] == True and column['TNcard'] == False:
        return 'cash'
    elif column['TNcash'] == False and column['TNcard'] == True:
        return 'card'
    elif column['TNcash'] == False and column['TNcard'] == False:
        return 'missing'
    elif column['TNcash'] == True and column['TNcard'] == True:
        return 'missing'
    else:
        return 'unknown'

# ใช้ .apply() เพื่อสร้างคอลัมน์ใหม่
P2019Fnodrop['Payment method'] = P2019F.apply(combine_values, axis=1)
print(P2019F)
````

````python
P2019F.reset_index()
````

````python
P2019Fnodrop.drop(['BeginDateTime_x','EndDateTime','TranTime','BreakTime', 'BeginDateTime_y','WorkstationGroupID_y'], axis=1,inplace=True)
P2019Fnodrop
````

````python
P2019F.to_csv('P2019F.csv', index=False)
````
